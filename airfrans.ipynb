```python
import numpy as np
import matplotlib.pyplot as plt

import torch
import torch.nn as nn
from torch.utils.data import TensorDataset, DataLoader

from sklearn.model_selection import train_test_split

from pinn import PINN
from physics import (
    navier_stokes_residual,
    physics_loss,
)

device = torch.device(
    "cuda" if torch.cuda.is_available() else "cpu"
)

print(device)
```

```python
#################################################
# Merge all simulations
#################################################

all_points = np.concatenate(dataset_list, axis=0)

print(all_points.shape)
```

```python
#################################################
# Inputs
#################################################

X = all_points[:, :7]

#################################################
# Targets
#################################################

Y = all_points[:, 7:10]

print("Input :", X.shape)
print("Target:", Y.shape)
```

```python
#################################################
# Random sample
#################################################

N = 50000

idx = np.random.choice(
    len(X),
    N,
    replace=False
)

X = X[idx]
Y = Y[idx]

print(X.shape)
```

```python
X_train, X_val, Y_train, Y_val = train_test_split(
    X,
    Y,
    test_size=0.2,
    random_state=42
)

print(X_train.shape)
print(X_val.shape)
```

```python
train_dataset = TensorDataset(
    torch.FloatTensor(X_train),
    torch.FloatTensor(Y_train),
)

val_dataset = TensorDataset(
    torch.FloatTensor(X_val),
    torch.FloatTensor(Y_val),
)

train_loader = DataLoader(
    train_dataset,
    batch_size=1024,
    shuffle=True,
)

val_loader = DataLoader(
    val_dataset,
    batch_size=1024,
)
```

```python
model = PINN(
    input_dim=7,
    output_dim=3,
    hidden_dim=128,
    num_hidden=8,
    activation="tanh",
).to(device)

optimizer = torch.optim.Adam(
    model.parameters(),
    lr=1e-3,
)

criterion = nn.MSELoss()
```

```python
##############################################################
# PINN Training
##############################################################

epochs = 100

lambda_data = 1.0
lambda_physics = 0.1

best_loss = float("inf")

history = {
    "train": [],
    "val": [],
}

for epoch in range(epochs):

    ##########################################################
    # TRAIN
    ##########################################################

    model.train()

    running_loss = 0.0

    for x_batch, y_batch in train_loader:

        x_batch = x_batch.to(device)
        y_batch = y_batch.to(device)

        optimizer.zero_grad()

        ######################################################
        # Physics
        ######################################################

        continuity, mx, my, u_pred, v_pred, p_pred = \
            navier_stokes_residual(
                model,
                x_batch,
            )

        ######################################################
        # Data
        ######################################################

        prediction = torch.cat(
            [
                u_pred,
                v_pred,
                p_pred,
            ],
            dim=1,
        )

        data_loss = criterion(
            prediction,
            y_batch,
        )

        ######################################################
        # Physics Loss
        ######################################################

        p_loss = physics_loss(
            continuity,
            mx,
            my,
        )

        ######################################################
        # Total Loss
        ######################################################

        loss = (
            lambda_data * data_loss
            + lambda_physics * p_loss
        )

        loss.backward()

        optimizer.step()

        running_loss += loss.item()

    train_loss = running_loss / len(train_loader)

    ##########################################################
    # VALIDATION
    ##########################################################

    model.eval()

    running_val = 0.0

    with torch.no_grad():

        for x_batch, y_batch in val_loader:

            x_batch = x_batch.to(device)
            y_batch = y_batch.to(device)

            u, v, p = model(x_batch)

            pred = torch.cat(
                [
                    u,
                    v,
                    p,
                ],
                dim=1,
            )

            val_loss = criterion(
                pred,
                y_batch,
            )

            running_val += val_loss.item()

    val_loss = running_val / len(val_loader)

    history["train"].append(train_loss)
    history["val"].append(val_loss)

    ##########################################################
    # SAVE BEST
    ##########################################################

    if val_loss < best_loss:

        best_loss = val_loss

        torch.save(
            model.state_dict(),
            "best_pinn.pt",
        )

    ##########################################################
    # LOG
    ##########################################################

    print(
        f"Epoch {epoch+1:03d} | "
        f"Train {train_loss:.6f} | "
        f"Val {val_loss:.6f}"
    )
```

```python
plt.figure(figsize=(8,5))

plt.plot(
    history["train"],
    label="Train",
)

plt.plot(
    history["val"],
    label="Validation",
)

plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()

plt.grid(True)

plt.show()
```
